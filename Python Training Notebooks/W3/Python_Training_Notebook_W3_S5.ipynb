# Updating df using np.where 
#### NumPy is a library for the Python programming language, adding support for large, multi-dimensional arrays and matrices, along with a large collection of high-level mathematical functions to operate on these arrays.

```python
import numpy as np
```

### Updating 'Gross' column for 'Tumbbad' using np.where
Sample syntax:  
df['col'] = np.where(filter_condition, x, y)   
value 'x' is assigned to the column if the rows satisfies the filter_condition, 'y' otherwise

```python
imdb_df['Gross'] = np.where(imdb_df['Title']=='Tumbbad', '$800M',imdb_df['Gross'])
imdb_df
```

### Updating 'Gross' column for  multiple rows using np.where

```python
imdb_df['Gross'] = np.where(imdb_df['Title']=='The Shining', '$5M',
                            (np.where(imdb_df['Title']=='Troll 2', '$1.03M', imdb_df['Gross'])))
```

```python
imdb_df
```

### Concatenating dataframes

```python
# Creating a dataframe with 2 rows which has a similar structure as that of imdb_df
dict1 = {'Title': ['Saw 2','Saw 3'],
        'Year': ['2005','2006'],
        'Runtime': [105, 108],
        'Genre': ['Horror','Thriller'],
        'Rating': [8,7],
        'Director': 'James Wan',
        'Votes': [435678,435679],
        'Gross': ['$100M',np.NaN]}
df1 = pd.DataFrame(dict1)
df1
```

### concat -  joins two or more dataframes (join - inner, full, left, right) based on index value along the specified axis (axis=0 (row) / axis=1 (column))
#### Syntax:  
pandas.concat(objs, *, axis=0, join='outer', ignore_index=False,   
              keys=None, levels=None, names=None, verify_integrity=False, sort=False, copy=None)

```python
pd.concat([imdb_df,df1])  # refer to the above syntax for default param values
```

```python
imdb_df[imdb_df['Title']=='Hostel']
```

```python
dict2 = {'Title': ['Hostel 2','Hostel 3'],
        'Year': ['2005','2006'],
        'Runtime': [105, 108],
        'Genre': ['Horror','Horror'],
        'Rating': [8,7],
        'Director': 'Unknown',
        'Votes': [435678,435679],
        'Gross': ['$900M','$600M']}
df2 = pd.DataFrame(dict2)
df2
```

```python
pd.concat([imdb_df,df1,df2])
```

```python
"""
ignore_index = True will not retain the original index from df1 and df2 
and assign consecutive index values based on the first df (imbd_df)
"""
pd.concat([imdb_df,df1,df2],ignore_index=True)
```
