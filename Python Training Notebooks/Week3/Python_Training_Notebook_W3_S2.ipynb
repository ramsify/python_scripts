```python
auto_csv = r"C:\Users\ramananb\Downloads\Automobile.csv" 
df = pd.read_csv(auto_csv)
df
```

```python
lst = ['datsun pl510','amc gremlin','chevrolet impala','buick century']
```

```python
# passing the above list to slice the dataframe in the 'name' columns
df2 = df[df['name'].isin(lst)]
```

```python
df2
```

```python
# using list negation to create another dataframe that doesn't contain any of the list names in 'name' column
df3 = df[~df['name'].isin(lst)]
```

```python
df3
```

```python
# here if you see the df2 is a slice of the actual df and it contains the exact index values from the original data frame. 
# Now we can use reset_index to reorder the index and we dont want the old values so we use 'drop=True' parameter
df2
```

```python
df2.reset_index(drop=True,inplace=True)
```

```python
df2
```

```python
df3.reset_index(drop=True,inplace=True)
df3
```
